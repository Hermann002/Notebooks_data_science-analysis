## Preprocessing

```python
train_data_ = train_data_[full_join]
test_data_ = test[full_join]
test_data_.head()
```

### standardisation des variables

```python
numeric_features = train_data_.select_dtypes({'int64', 'float64'}).columns
numeric_features
```

```python
from sklearn.preprocessing import StandardScaler, OneHotEncoder

scaler = StandardScaler()
train_scl = scaler.fit_transform(train_data_[numeric_features])
test_scl = scaler.fit_transform(test_data_[numeric_features])

train_scl
```

```python
train_X_num = pd.DataFrame(train_scl, columns=numeric_features)
test_X_num = pd.DataFrame(test_scl, columns=numeric_features)
```

```python
train_X_num.head()
```

### Encode Categorical features

```python
encoder = OneHotEncoder(sparse_output=False)
train_enc = encoder.fit_transform(train_data_[categorical_features])
test_enc = encoder.fit_transform(test_data_[categorical_features])
```

```python
encoder_columns = encoder.get_feature_names_out(train_data_[categorical_features].columns)
train_X_cat = pd.DataFrame(train_enc, columns=encoder_columns)
test_X_cat = pd.DataFrame(test_enc, columns=encoder_columns)
train_X_cat.head()
```

```python
train_X = pd.concat([train_X_num, train_X_cat], axis=1)
test_X = pd.concat([test_X_num, test_X_cat], axis=1)
train_X.head()
```

### Traitement des valeurs manquantes

```python
train_X.info()
```

```python
from sklearn.impute import KNNImputer

imputer = KNNImputer(n_neighbors=3)
train_X_imp = imputer.fit_transform(train_X)
train_X_imp = pd.DataFrame(train_X_imp, columns=train_X.columns)
train_X_imp.info()
```

```python
test_X_imp = imputer.fit_transform(test_X)
test_X_imp = pd.DataFrame(test_X_imp, columns=test_X.columns)
test_X_imp.info()
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.linear_model import LinearRegression
from xgboost import XGBRegressor, XGBRFRegressor
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.metrics import root_mean_squared_error
## set linear regression

```

```python
X_train, X_test, y_train,  y_test = train_test_split(train_X_imp, TARGET, test_size=0.2, random_state=42)

print(X_train.shape)
print(y_train.shape)
print(X_test.shape)
print(y_test.shape)
```

```python
model_RFR = RandomForestRegressor()
model_RFR.fit(X_train, y_train)
y_preds = model_RFR.predict(X_test)
RMSE = root_mean_squared_error(y_test, y_preds)

print(RMSE)
```

```python
model_XGBR = XGBRegressor()
model_XGBR.fit(X_train, y_train)
y_preds = model_XGBR.predict(X_test)
RMSE = root_mean_squared_error(y_test, y_preds)

print(RMSE)
```

```python
model_XGBRFR = XGBRFRegressor()
model_XGBRFR.fit(X_train, y_train)
y_preds = model_XGBRFR.predict(X_test)
RMSE = root_mean_squared_error(y_test, y_preds)

print(RMSE)
```
